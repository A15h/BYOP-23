```python
name = "Salman Khan"
```

```python
# function to create a dictionary oftitle and sentiment
def get_data(keyword):
  news = []
  allarticles = newsapi.get_everything(
    q=keyword,
    language='en',   
  )
  articles = allarticles["articles"]
  for i in articles:
    article = {'title':i["title"],'link':i['url'],'published':i['publishedAt']}
    news.append(article)
  return news

data = get_data(name)
data
```
